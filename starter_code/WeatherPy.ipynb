```python
import pandas as pd
import gmaps
from pprint import pprint
import matplotlib.pyplot as plt
from api_keys import weather_api_key
import requests
import json
from citipy import citipy
import numpy as np
from scipy import stats
```

## Generate Cities List

```python
city=[]
lat_lon=[]
new_lat=[]
new_lon=[]

# Generate latitudes(-90 to 90) and longitudes(-180 to 180) using random number generator and multiply by 100

lat=np.random.uniform(-90,90,1500)
lon=np.random.uniform(-180,180,1500)

lat_lon=zip(lat,lon)

for coor in lat_lon:
    new_city=citipy.nearest_city(coor[0],coor[1])
    name=new_city.city_name
    if name not in city:
        city.append(name)
        new_lat.append(coor[0])
        new_lon.append(coor[1])

city_df=pd.DataFrame({
    'City':city,
    'Latitude':new_lat,
    'Longitude':new_lon
})

city_df
    
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
city=[]
country=[]
date=[]
lat=[]
lon=[]
temp=[]
humidity=[]
cloudiness=[]
wind_speed=[]
index_count=0
set_count=1
units = 'metric'

url='http://api.openweathermap.org/data/2.5/weather?'
query_url = f'{url}appid={weather_api_key}&units={units}&q='

for index,row in city_df.iterrows():
    try:
        response=requests.get(query_url+row['City']).json()
        city.append(response['name'])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        lat.append(response['coord']['lat'])
        lon.append(response['coord']['lon'])
        temp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
        if index_count>=50:
            index_count=0
            set_count+=1
            
        else:
            index_count+=1
        
        print(f'Processing Record {index_count} of Set {set_count} | {response["name"]}')
        
            
        
    except:
        print('City not found. Skipping...')
    
    
        
    
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
new_city_df=pd.DataFrame({
    'city':city,
    'country':country,
    'date':date,
    'lat':lat,
    'lon':lon,
    'max_temp':temp,
    'humidity':humidity,
    'cloudiness':cloudiness,
    'wind_speed':wind_speed
})


new_city_df.to_csv('output.csv', index=False)
new_city_df
```

## Linear Regression

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
northern_hem_df=new_city_df.loc[new_city_df['lat']>0]
# northern_hem_df.info()

x_values=northern_hem_df['lat']
y_values=northern_hem_df['max_temp']
plt.scatter(x_values,y_values,color='purple',alpha=1)
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
print(regress_values)

plt.plot(x_values,regress_values,"y-")
plt.title('Northern Hemisphere - Max Temp vs. Latitude Linear Regression')
plt.xlabel('Latitude')
plt.ylabel('Max temp')
plt.annotate(line_eq,(10,5),fontsize=15,color="blue")
plt.savefig('Northern Hemisphere - Max Temp vs. Latitude Linear Regression.png')
```
